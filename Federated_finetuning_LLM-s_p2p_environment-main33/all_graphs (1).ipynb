```python
import numpy as np
import matplotlib.pyplot as plt
def addlabels1(x,y1):
  for i in range(len(x)):
    plt.text(i,y1[i]+0.5,y1[i],ha='center',fontweight ='bold', fontsize = 12)
def addlabels2(x,y1):
  for i in range(len(x)):
    plt.text(i+0.2,y1[i]+0.5,y1[i],ha='center',fontweight ='bold', fontsize = 12)

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
SERVER = [38, 41.8, 45.4]
SERVERLESS = [65, 40, 41.5]

# Set position of bar on X axis
br1 = np.arange(len(SERVER))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
addlabels1(br1,SERVER)
addlabels2(br1,SERVERLESS)
# Make the plot
plt.bar(br1, SERVER, color ='lightblue', width = barWidth,
		edgecolor ='grey', label ='Centralized')
plt.bar(br2, SERVERLESS, color ='lightpink', width = barWidth,
		edgecolor ='grey', label ='Fully decentralized')

# Adding Xticks
plt.xlabel('Number of workers', fontweight ='bold', fontsize = 15)
plt.ylabel('Latency (Min)', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(SERVER))],
		['5', '10', '20'])
plt.title("Latency - All cases",fontweight ='bold', fontsize = 15)

plt.legend()
plt.savefig('Latency.jpg', dpi=600)  # Adjust the DPI value as needed
plt.show()
```
