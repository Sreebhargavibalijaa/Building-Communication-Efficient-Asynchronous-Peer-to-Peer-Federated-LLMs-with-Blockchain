```python
import numpy as np
import matplotlib.pyplot as plt
def addlabels1(x,y1):
  for i in range(len(x)):
    plt.text(i,y1[i]+0.89,y1[i],ha='center',fontweight ='bold', fontsize = 12)
def addlabels2(x,y1):
  for i in range(len(x)):
    plt.text(i+0.23,y1[i]+0.89,y1[i],ha='center',fontweight ='bold', fontsize = 12)

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
SERVER = [280.557, 628, 810]
SERVERLESS = [105, 122, 187]

# Set position of bar on X axis
br1 = np.arange(len(SERVER))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
addlabels1(br1,SERVER)
addlabels2(br1,SERVERLESS)
# Make the plot
plt.bar(br1, SERVER, color ='lightblue', width = barWidth,
		edgecolor ='grey', label ='Centralized')
plt.bar(br2, SERVERLESS, color ='lightpink', width = barWidth,
		edgecolor ='grey', label ='Fully decentralized')

# Adding Xticks
plt.xlabel('Number of workers', fontweight ='bold', fontsize = 15)
plt.ylabel('Latency (Min)', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(SERVER))],
		['5', '10', '20'])
plt.title("Latency - All cases",fontweight ='bold', fontsize = 15)

plt.legend()
plt.savefig('Latency.jpg', dpi=600)  # Adjust the DPI value as needed
plt.show()
```

The above graph shows the latency comparison of clients with change in workers. Here average latency is the total time required to compute the global model for each client. We have observed the average latency of clients has increased with the number of workers because of an increase in communication updates.

Information passing time form central node to remaining nodes with out async blockchain

Calculation Process of Information processing time

Synchronous way
Information passing time from node1 to all nodes = time(1-2) +  time(1-3) + time(1-4) + time(1-5)+ time(1-6)+ time(1-7)+ time(1-8)+ time(1-9) + time(1-10)

ASynchronous way
Information passing time from node1 to all nodes = Max(time(1-2),time(1-3),time(1-4),time(1-5),time(1-6),time(1-7),time(1-8),time(1-9),time(1-10))

model = 0.40362595301121473 GB
channel bandwidth(1-2) = 145 mbps
Time(1-2) = model size / channel bandwidth(1-2) = 0.4036/145 = 2.7 s

Time(1-3) = model size / channel bandwidth(1-3) = 0.4036/343 = 1.17 s

Time(1-4), shortest path from 1 to 4 is from 1-3 and 3-4

Time(1-4) = model size / channel bandwidth(1-3)+ model size / channel bandwidth(3-4)

Synchronous Information passing time from node1 to all nodes = time(1-2) +  time(1-3) + time(1-4) + time(1-5)+ time(1-6)+ time(1-7)+ time(1-8)+ time(1-9) + time(1-10)

Information passing time from node1 to all nodes = 2.7+1.17 .... = 44.8sec


ASynchronous Information passing time from node1 to all nodes = Max(time(1-2),time(1-3),time(1-4),time(1-5),time(1-6),time(1-7),time(1-8),time(1-9),time(1-10))

ASynchronous Information passing time from node1 to all nodes = Max( 2.7, 1.17,4.5,9.38 .... 1,0.89 ...)= 9.38 sec


```python
import numpy as np
import matplotlib.pyplot as plt
def addlabels1(x,y1):
  for i in range(len(x)):
    plt.text(i,y1[i]+0.2,y1[i],ha='center',fontweight ='bold', fontsize = 12)
def addlabels2(x,y1):
  for i in range(len(x)):
    plt.text(i+0.2,y1[i]+0.2,y1[i],ha='center',fontweight ='bold', fontsize = 12)
# set width of bar
barWidth = 0.2
fig = plt.subplots(figsize =(12, 8))

# set height of bar
Synchronous = [44.64,35.5,27.5]
Asynchronous = [9.3,9.3,7.44]

# Set position of bar on X axis
br1 = np.arange(len(Synchronous))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
addlabels1(br1,Synchronous)
addlabels2(br1,Asynchronous)
# Make the plot
plt.bar(br1, Synchronous, color ='lightblue', width = barWidth,
		edgecolor ='grey', label ='Synchronous')
plt.bar(br2, Asynchronous, color ='lightyellow', width = barWidth,
		edgecolor ='grey', label ='Asynchronous')

# Adding Xticks
plt.xlabel('Anomaly detection methods', fontweight ='bold', fontsize = 15)
plt.ylabel('Information passing time (Sec)', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Asynchronous))],
		['DBSCAN', 'Modified Z- scores', 'Pagerank'])
plt.title("Information passing time, without async blockchain", fontweight ='bold', fontsize = 15)

plt.legend()
plt.savefig('Information_passing_time_with_out_blockchain.jpg', dpi=600)  # Adjust the DPI value as needed
plt.show()
```

The above graph shows the time required for passing the information from first client to all clients for 10 clients in asynchronous and synchronous case which is further based on channel bandwidth and model size, in the Synchronous mechanism
Later we have removed the anomaly workers based on different anomaly detection methods and recalculated the information passing time, we have observed that information passing time is less for PageRankmethod compared to other methods
